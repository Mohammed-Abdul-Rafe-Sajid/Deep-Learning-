# Comparison of Optimization Algorithms

## Dataset

Any standard dataset can be used for comparison:

- MNIST Digit Dataset  
- CIFAR-10  
- Iris Dataset  

These datasets are commonly used to evaluate the performance of optimization algorithms.

---

## 1. Stochastic Gradient Descent (SGD)

### Description
Stochastic Gradient Descent updates the model parameters using one training sample at a time.

### Key Features
- Updates weights for each training sample
- Simple and easy to implement
- Computationally efficient per update

### Advantages
- Low memory usage
- Suitable for large datasets
- Can escape local minima due to noisy updates

### Disadvantages
- Slow convergence
- High variance in updates
- Can oscillate during training

---

## 2. Momentum Based Gradient Descent

### Description
Momentum-based Gradient Descent accelerates SGD by using past gradients to smooth updates.

### Key Features
- Uses previous gradients to influence current updates
- Reduces oscillations
- Speeds up convergence

### Advantages
- Faster convergence than SGD
- More stable updates
- Effective in deep or narrow valleys

### Disadvantages
- Requires tuning of the momentum parameter
- Slightly higher computation than SGD

---

## 3. Adam Optimizer

### Description
Adam (Adaptive Moment Estimation) combines Momentum and Adaptive Learning Rate methods.

### Key Features
- Maintains moving averages of gradients and squared gradients
- Automatically adapts learning rates for each parameter
- Widely used in deep learning

### Advantages
- Very fast convergence
- Handles sparse gradients well
- Minimal hyperparameter tuning

### Disadvantages
- Higher memory consumption
- May generalize worse than SGD in some cases

---

## Performance Comparison

| Optimizer | Convergence Speed | Stability | Oscillation | Tuning Required |
|----------|------------------|-----------|-------------|----------------|
| SGD | Slow | Low | High | High |
| Momentum GD | Medium to Fast | Medium | Low | Medium |
| Adam | Fast | High | Very Low | Low |

---

## Conclusion

- SGD is simple but may converge slowly and oscillate.
- Momentum-based Gradient Descent improves speed and stability.
- Adam Optimizer provides fast convergence and is widely used in deep learning models.

Adam generally performs best on complex datasets such as MNIST and CIFAR-10.


### 1. IMPORT LIBRARIES


```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
```

## 2. Load and preprocess data

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
x_train=x_train/255.0
x_test=x_test/255.0
```

## Step 3: Create Neural Network Model

```python
def create_model():
  model=keras.Sequential(
      [
          layers.Flatten(input_shape=(28,28)),
          layers.Dense(128,activation='relu'),
          layers.Dense(10,activation='softmax')
      ]
  )
  return model
```

## Step 4: Train Model Using SGD


```python
model_sgd=create_model()
model_sgd.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01,momentum=0.0,nesterov=False),
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
history_sgd=model_sgd.fit(x_train,y_train,epochs=10,validation_split=0.2)
```

## Step 5: Train Model Using Momentum Based GD

```python

model_momentum = create_model()
model_momentum.compile(
 optimizer=keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=False),
 loss='sparse_categorical_crossentropy',
 metrics=['accuracy']
)
history_momentum = model_momentum.fit(x_train, y_train, epochs=10,
validation_split=0.2)

```

## Step 6: Train Model Using Adam Optimizer

```python
model_adam = create_model()
model_adam.compile(
 optimizer=keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999,
epsilon=1e-7, amsgrad=False),
 loss='sparse_categorical_crossentropy',
 metrics=['accuracy']
)
history_adam = model_adam.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

## Step 7: Plot Accuracy Comparison

```python

plt.plot(history_sgd.history['val_accuracy'], label='SGD')
plt.plot(history_momentum.history['val_accuracy'], label='Momentum')
plt.plot(history_adam.history['val_accuracy'], label='Adam')
plt.title("Optimizer Accuracy Comparison")
plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.show()
```

## Step 8: Plot Loss Comparison

```python

plt.plot(history_sgd.history['val_loss'], label='SGD')
plt.plot(history_momentum.history['val_loss'], label='Momentum')
plt.plot(history_adam.history['val_loss'], label='Adam')
plt.title("Optimizer Loss Comparison")
plt.xlabel("Epochs")
plt.ylabel("Validation Loss")
plt.legend()
plt.show()
```
