```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Define input sequences
X = np.array([
    [1,2,3],
    [2,3,4],
    [3,4,5],
    [4,5,6],
    [5,6,7]
])

# Define target outputs
y = np.array([4,5,6,7,8])

# Reshape input data to fit RNN requirements (samples, timesteps, features)
# X.shape[0] is the number of samples (5)
# X.shape[1] is the number of timesteps (3)
# 1 is the number of features for each timestep
X = X.reshape((X.shape[0], X.shape[1], 1))

# Define the SimpleRNN model
model = Sequential()
# Add a SimpleRNN layer with 10 units, tanh activation, and specify input shape
# input_shape=(3,1) means 3 timesteps and 1 feature per timestep
model.add(SimpleRNN(10, activation='tanh', input_shape=(3,1)))
# Add a Dense output layer with 1 unit for regression
model.add(Dense(1))

# Compile the model
# Use 'adam' optimizer and 'mean squared error' (mse) loss for regression tasks
model.compile(optimizer='adam', loss='mse')

# Print a summary of the model's architecture
model.summary()

# Train the model
# X: input features
# y: target labels
# epochs: number of times to iterate over the entire dataset
# verbose=1: show progress bar during training
model.fit(X, y, epochs=500, verbose=1)

# Prepare new input for prediction
# This is the sequence [5,6,7] reshaped to (1, 3, 1) to match model input requirements
test_input = np.array([[5,6,7]])
test_input = test_input.reshape((1,3,1))

# Make a prediction using the trained model
prediction = model.predict(test_input)

# Print the predicted value
print("Predicted value:", prediction)
```
