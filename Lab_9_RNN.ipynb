```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Define input sequences
X = np.array([
    [1,2,3],
    [2,3,4],
    [3,4,5],
    [4,5,6],
    [5,6,7]
])

# Define target outputs
y = np.array([4,5,6,7,8])

# Reshape input data to fit RNN requirements (samples, timesteps, features)
# X.shape[0] is the number of samples (5)
# X.shape[1] is the number of timesteps (3)
# 1 is the number of features for each timestep
X = X.reshape((X.shape[0], X.shape[1], 1))

# Define the SimpleRNN model
model = Sequential()
# Add a SimpleRNN layer with 10 units, tanh activation, and specify input shape
# input_shape=(3,1) means 3 timesteps and 1 feature per timestep
model.add(SimpleRNN(10, activation='tanh', input_shape=(3,1)))
# Add a Dense output layer with 1 unit for regression
model.add(Dense(1))

# Compile the model
# Use 'adam' optimizer and 'mean squared error' (mse) loss for regression tasks
model.compile(optimizer='adam', loss='mse')

# Print a summary of the model's architecture
model.summary()

# Train the model
# X: input features
# y: target labels
# epochs: number of times to iterate over the entire dataset
# verbose=1: show progress bar during training
model.fit(X, y, epochs=500, verbose=1)

# Prepare new input for prediction
# This is the sequence [5,6,7] reshaped to (1, 3, 1) to match model input requirements
test_input = np.array([[5,6,7]])
test_input = test_input.reshape((1,3,1))

# Make a prediction using the trained model
prediction = model.predict(test_input)

# Print the predicted value
print("Predicted value:", prediction)
```

```python
# This cell contains the entire concept and code for training a text generation model
# using an FAQ dataset, with detailed comments for better comprehension.

# 1. Define the FAQ text (This is the input data for our model)
faqs = """About the Program
What is the course fee for  Data Science Mentorship Program (DSMP 2023)
The course follows a monthly subscription model where you have to make monthly payments of Rs 799/month.
What is the total duration of the course?
The total duration of the course is 7 months. So the total course fee becomes 799*7 = Rs 5600(approx.)
What is the syllabus of the mentorship program?
We will be covering the following modules:
Python Fundamentals
Python libraries for Data Science
Data Analysis
SQL for Data Science
Maths for Machine Learning
ML Algorithms
Practical ML
MLOPs
Case studies
You can check the detailed syllabus here - https://learnwith.campusx.in/courses/CampusX-Data-Science-Mentorship-Program-637339afe4b0615a1bbed390
Will Deep Learning and NLP be a part of this program?
No, NLP and Deep Learning both are not a part of this programвҖҷs curriculum.
What if I miss a live session? Will I get a recording of the session?
Yes all our sessions are recorded, so even if you miss a session you can go back and watch the recording.
Where can I find the class schedule?
Checkout this google sheet to see month by month time table of the course - https://docs.google.com/spreadsheets/d/16OoTax_A6ORAeCg4emgexhqqPv3noQPYKU7RJ6ArOzk/edit?usp=sharing.
What is the time duration of all the live sessions?
Roughly, all the sessions last 2 hours.
What is the language spoken by the instructor during the sessions?
Hinglish
How will I be informed about the upcoming class?
You will get a mail from our side before every paid session once you become a paid user.
Can I do this course if I am from a non-tech background?
Yes, absolutely.
I am late, can I join the program in the middle?
Absolutely, you can join the program anytime.
If I join/pay in the middle, will I be able to see all the past lectures?
Yes, once you make the payment you will be able to see all the past content in your dashboard.
Where do I have to submit the task?
You donвҖҷt have to submit the task. We will provide you with the solutions, you have to self evaluate the task yourself.
Will we do case studies in the program?
Yes.
Where can we contact you?
You can mail us at nitish.campusx@gmail.com
Payment/Registration related questions
Where do we have to make our payments? Your YouTube channel or website?
You have to make all your monthly payments on our website. Here is the link for our website - https://learnwith.campusx.in/
Can we pay the entire amount of Rs 5600 all at once?
Unfortunately no, the program follows a monthly subscription model.
What is the validity of monthly subscription? Suppose if I pay on 15th Jan, then do I have to pay again on 1st Feb or 15th Feb
15th Feb. The validity period is 30 days from the day you make the payment. So essentially you can join anytime you donвҖҷt have to wait for a month to end.
What if I donвҖҷt like the course after making the payment. What is the refund policy?
You get a 7 days refund period from the day you have made the payment.
I am living outside India and I am not able to make the payment on the website, what should I do?
You have to contact us by sending a mail at nitish.campusx@gmail.com
Post registration queries
Till when can I view the paid videos on the website?
This one is tricky, so read carefully. You can watch the videos till your subscription is valid. Suppose you have purchased subscription on 21st Jan, you will be able to watch all the past paid sessions in the period of 21st Jan to 20th Feb. But after 21st Feb you will have to purchase the subscription again.
But once the course is over and you have paid us Rs 5600(or 7 installments of Rs 799) you will be able to watch the paid sessions till Aug 2024.
Why lifetime validity is not provided?
Because of the low course fee.
Where can I reach out in case of a doubt after the session?
You will have to fill a google form provided in your dashboard and our team will contact you for a 1 on 1 doubt clearance session
If I join the program late, can I still ask past week doubts?
Yes, just select past week doubt in the doubt clearance google form.
I am living outside India and I am not able to make the payment on the website, what should I do?
You have to contact us by sending a mail at nitish.campusx@gmai.com
Certificate and Placement Assistance related queries
What is the criteria to get the certificate?
There are 2 criterias:
You have to pay the entire fee of Rs 5600
You have to attempt all the course assessments.
I am joining late. How can I pay payment of the earlier months?
You will get a link to pay fee of earlier months in your dashboard once you pay for the current month.
I have read that Placement assistance is a part of this program. What comes under Placement assistance?
This is to clarify that Placement assistance does not mean Placement guarantee. So we dont guarantee you any jobs or for that matter even interview calls. So if you are planning to join this course just for placements, I am afraid you will be disappointed. Here is what comes under placement assistance
Portfolio Building sessions
Soft skill sessions
Sessions with industry mentors
Discussion on Job hunting strategies
"""

# 2. Import necessary libraries for building and training the model
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np # Required for numerical operations like argmax

# 3. Initialize and fit the Tokenizer
# The Tokenizer processes text data by converting words into numerical indices.
# It creates a vocabulary from the provided text.
tokenizer = Tokenizer()
tokenizer.fit_on_texts([faqs])

# Calculate the vocabulary size. We add 1 because word indices start from 1, and 0 is reserved for padding.
vocab_size = len(tokenizer.word_index) + 1

# 4. Prepare input sequences for training
# We will create n-gram sequences from each sentence in the FAQs.
# For example, if a sentence is "what is the fee", it will generate sequences like:
# [what, is]
# [what, is, the]
# [what, is, the, fee]
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1, len(tokenized_sentence)): # Start from 1 to ensure at least two words in a sequence
    input_sequences.append(tokenized_sentence[:i+1])

# 5. Pad the sequences to ensure uniform length
# Neural networks require input sequences to have the same length.
# We find the maximum length among all generated sequences.
max_len = max([len(x) for x in input_sequences])

# We pad the sequences with zeros at the beginning ('pre' padding) to match 'max_len'.
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')

# 6. Split data into input (X) and output (y)
# For each padded sequence, the input X will be all words except the last one.
# The output y will be the last word in the sequence, which the model needs to predict.
X = padded_input_sequences[:, :-1]
y = padded_input_sequences[:, -1]

# 7. Convert the target variable (y) to one-hot encoding
# One-hot encoding is necessary when using 'categorical_crossentropy' loss for multi-class classification.
# Each word index in y is converted into a vector of zeros with a '1' at the index corresponding to the word.
y = to_categorical(y, num_classes=vocab_size)

# 8. Define the Neural Network Model (LSTM-based)
model = Sequential()

# Embedding Layer:
# Converts numerical word indices into dense vectors of a fixed size (100 dimensions here).
# This helps the model understand semantic relationships between words.
# `input_length` is inferred from the input data (X.shape[1]), so it's not strictly necessary to provide it directly anymore,
# but here it's explicitly set to `max_len - 1` which is the length of X.
model.add(Embedding(vocab_size, 100, input_length=X.shape[1]))

# First LSTM Layer:
# Long Short-Term Memory (LSTM) layers are particularly good at processing sequences.
# `return_sequences=True` is crucial here because we are stacking another LSTM layer on top.
# It ensures that the first LSTM layer outputs a sequence of its own outputs, not just the final output.
model.add(LSTM(150, return_sequences=True))

# Second LSTM Layer:
# This layer processes the sequence output from the first LSTM layer.
# It does not need `return_sequences=True` as it's the last LSTM layer before the Dense output layer.
model.add(LSTM(150))

# Dense Output Layer:
# This layer has `vocab_size` units, one for each word in our vocabulary.
# The 'softmax' activation function ensures that the output is a probability distribution over all words,
# indicating the likelihood of each word being the next word in the sequence.
model.add(Dense(vocab_size, activation='softmax'))

# 9. Compile the Model
# 'categorical_crossentropy' is the loss function used for multi-class classification with one-hot encoded labels.
# 'adam' is a widely used and effective optimizer.
# 'accuracy' is included as a metric to monitor training progress.
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Display a summary of the model's architecture, including the number of parameters.
model.summary()

# 10. Train the Model
# The model learns to predict the next word based on the preceding words over multiple epochs.
# An epoch is one complete pass through the entire training dataset.
print("\nTraining the model...")
model.fit(X, y, epochs=50, verbose=0) # verbose=1 shows a progress bar during training

# 11. Function to Generate Text (Demonstration of Model Usage)
# This function takes a trained model, tokenizer, and a starting text to predict subsequent words.
def generate_text(model, tokenizer, max_sequence_len, seed_text, num_words_to_generate):
    generated_text = seed_text
    for _ in range(num_words_to_generate):
        # Tokenize the current generated text (seed_text + predicted words)
        token_list = tokenizer.texts_to_sequences([generated_text])[0]

        # Pad the token list to match the input length required by the model (X.shape[1])
        padded_token_list = pad_sequences([token_list], maxlen=X.shape[1], padding='pre')

        # Predict the next word's probability distribution
        # verbose=0 suppresses the progress bar for prediction
        predicted_probs = model.predict(padded_token_list, verbose=0)

        # Get the index of the word with the highest probability
        predicted_word_index = np.argmax(predicted_probs)

        output_word = ""
        # Find the actual word corresponding to the predicted index
        for word, index in tokenizer.word_index.items():
            if index == predicted_word_index:
                output_word = word
                break
        # Append the predicted word to the generated text
        generated_text += " " + output_word
    return generated_text

# 12. Demonstrate Text Generation
# Define a starting phrase (seed text)
seed_text = "what is the course"

# Generate 5 additional words based on the trained model
generated_sequence = generate_text(model, tokenizer, max_len, seed_text, 5)
print(f"\nSeed Text: '{seed_text}'")
print(f"Generated Sequence: '{generated_sequence}'")

seed_text_2 = "Can I join"
generated_sequence_2 = generate_text(model, tokenizer, max_len, seed_text_2, 5)
print(f"\nSeed Text: '{seed_text_2}'")
print(f"Generated Sequence: '{generated_sequence_2}'")
```
